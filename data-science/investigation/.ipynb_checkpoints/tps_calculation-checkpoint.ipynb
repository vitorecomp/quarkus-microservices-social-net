```python
print("Tps by cpu")
    
import warnings
warnings.filterwarnings("ignore")

import numpy as np
import math
import matplotlib.pyplot as plt

from sklearn.linear_model import Ridge
from sklearn.preprocessing import PolynomialFeatures
from sklearn.preprocessing import SplineTransformer
from sklearn.pipeline import make_pipeline


# generate points used to plot

# generate points and keep a subset of them
x_old = [0, 1000, 2000, 3000, 4000, 5000, 6000, 7000, 8000, 9000, 11000]
# y = [0, 1, 1.8, 2.6, 3.2, 4.1, 5, 3, 5, 5, 6]
y = [0, 100, 200, 250, 300, 320, 350, 400, 410, 390, 400]

mx = max(x_old)
my = max(y)

x_plot = np.linspace(100, mx, 100)

# for i in range(2):
#     if(mx+i*100 < mx*1.5):
#         x_old.append(mx+i*100)
#         y.append(my)

x = np.array(x_old)

# create matrix versions of these arrays
X = x[:, np.newaxis]
X_plot = x_plot[:, np.newaxis]

colors = ['teal', 'yellowgreen', 'gold']
lw = 2


for degree in range(10):
    if(degree %2 == 0):
        modelt = 'poly'
        model = make_pipeline(PolynomialFeatures(math.floor(degree/2)+2), Ridge())
    else:
        modelt = 'spline'
        model = make_pipeline(SplineTransformer(n_knots=math.floor(degree/2)+2, degree=math.floor(degree/2)+2), Ridge(alpha=1e-3))
    model.fit(X, y)
    y_plot = model.predict(X_plot)
    poison = False
    for value in y_plot:
        if(value > max(y)*1.1):
            poison = True
            break
        if(value < y[0]):
            poison = True
            break
    if(poison):
        continue
    plt.figure()
    plt.scatter(x, y, color='navy', s=30, marker='o', label="training points")
    plt.plot(x_plot, y_plot, color='teal', linewidth=lw,label= modelt + " degree %d" % (math.floor(degree/2)+2))
    plt.legend(loc='lower left')
    # break

plt.show()
```
